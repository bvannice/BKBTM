```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("train.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
#Fill blank runtime entries with zero value
df["runtime"]=df["runtime"].fillna(0)
```

```python
df.info()
```

```python
# Change homepage column to yes or no format
df.loc[df["homepage"].notnull(),"homepage"]=1
df["homepage"]=df["homepage"].fillna(0)
```

```python
# Add column whether the movie has a tagline or not
df["Has_tagline"]=1
df.loc[df["tagline"].isnull(), "Has_tagline"]=0
```

```python
# Change budget, popularity, and runtime to percent values
#df['budget']=df['budget'].apply(lambda x:(x-np.min(df['budget']))/(np.max(df['budget']-np.min(df['budget']))))

#df['popularity']=df['popularity'].apply(lambda x:(x-np.min(df['popularity']))/(np.max(df['popularity']-np.min(df['popularity']))))

#df['runtime']=df['runtime'].apply(lambda x:(x-np.min(df['runtime']))/(np.max(df['runtime']-np.min(df['runtime']))))
```

```python
X=df[["budget", "popularity", "runtime", "homepage", "Has_tagline"]]
print(X.head())
```

```python
y = df["revenue"]
```

```python
print(X.shape, y.shape)
```

```python
#Train, test, split the train dataset
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
#Import RandomForest Regressor
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(n_estimators=200, n_jobs=1)
rf = rf.fit(X_train, y_train)
```

```python
rf.score(X_test, y_test)
```

```python
# Random Forests in sklearn will automatically calculate feature importance
importances = rf.feature_importances_
importances
```

```python
#Import Linear Regression model
from sklearn import linear_model
model = linear_model.LinearRegression()
```

```python
#fit and score the data
model.fit(X_train,y_train)
model.score(X_test,y_test)
```

```python
predict = rf.predict(X_test)
pd.DataFrame({"Prediction": predict.astype(int), "Actual": y_test})
```
