```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Read the csv file into a pandas DataFrame

movies = pd.read_csv('Resources/train.csv')
moviesTest = pd.read_csv('Resources/test.csv')

print(movies.shape, moviesTest.shape)
```

```python
movies.head()
```

```python
#homepage or no per Kyle H.
movies.loc[movies["homepage"].notnull(),"homepage"]=1
movies["homepage"]=movies["homepage"].fillna(0)

movies.loc[movies["belongs_to_collection"].notnull(),"belongs_to_collection"]=1
movies["belongs_to_collection"]=movies["belongs_to_collection"].fillna(0)

movies.loc[movies["production_companies"].notnull(),"production_companies"]=1
movies["production_companies"]=movies["production_companies"].fillna(0)

movies.loc[movies["tagline"].notnull(),"tagline"]=1
movies["tagline"]=movies["tagline"].fillna(0)

movies.loc[movies["poster_path"].notnull(),"poster_path"]=1
movies["poster_path"]=movies["poster_path"].fillna(0)

#get dummies
#moviesEnc = pd.get_dummies(movies, columns=["original_language"])
#moviesEnc.head(25)
```

```python
#Remove budget values less than 10
#movies = movies.replace(0, 1)
movies2 = movies[movies.budget > 10]

#movies2.sort_values(by="budget", ascending=True).head()
```

```python
movies2["runtime"].fillna(0)
movies3 = movies2[movies2.runtime > 0]

movies3.sort_values(by="budget", ascending=True).head()
```

```python
movies3.info()
```

```python
pd.set_option("display.max_columns", 58)
display(movies3)
```

```python
# budget vs revenue
plt.scatter(movies3.budget, movies3.revenue)
#save figure
plt.savefig("bud_v_rev")
```

```python
#Popularity vs Revenue
plt.scatter(movies3.popularity, movies3.revenue)
#save figure
plt.savefig("pop_v_rev")
```

```python
# Assign the data to X and y
X = movies3[["budget", "popularity", "homepage", "belongs_to_collection", "tagline", "production_companies", "runtime"]]
y = movies3["revenue"].values.reshape(-1, 1)
#X = movies.drop("popularity", axis=1)
#y = movies["popularity"]
print(X.shape, y.shape)
```

```python
#budget and popularity
movies3.sort_values(by="budget", ascending=False).head(10)[['title', 'budget', "popularity"]]
```

```python
#revenue
#g = movies.revenue
movies3.sort_values(by="revenue", ascending=False).head(10)[['title', 'revenue', "popularity"]]
```

```python
#movies = movies3.reset_index()
```

```python
# Split the data into training and testing
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=55)
```

```python
from sklearn.preprocessing import StandardScaler

# Create a StandardScater model and fit it to the training data
X_scaler = StandardScaler().fit(X_train)
y_scaler = StandardScaler().fit(y_train)
```

```python
# Transform the training and testing data using the X_scaler and y_scaler models

X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
y_train_scaled = y_scaler.transform(y_train)
y_test_scaled = y_scaler.transform(y_test)
```

```python
# Create a LinearRegression model and fit it to the scaled training data
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train_scaled, y_train_scaled)
```

```python
# Make predictions using the X_test_scaled data
# Plot y_test_scaled vs y_test_scaled
# Scatter plot y_test_scaled vs predictions

predictions = model.predict(X_test_scaled)
plt.scatter(model.predict(X_train_scaled), model.predict(X_train_scaled) - y_train_scaled, c="blue", label="Training Data")
plt.scatter(predictions, predictions - y_test_scaled, c="orange", label="Testing Data")
plt.legend()
plt.hlines(y=0, xmin=y_test_scaled.min(), xmax=y_test_scaled.max())
plt.title("Residual Plot")
plt.show()
#save figure
plt.savefig("bud_pop_vs_rev")
```

```python
# Used X_test_scaled, y_test_scaled, and model.predict(X_test_scaled) to calculate MSE and R2

from sklearn.metrics import mean_squared_error

MSE = mean_squared_error(y_test_scaled, predictions)
r2 = model.score(X_test_scaled, y_test_scaled)

print(f"MSE: {MSE}, R2: {r2}")


```

```python
#Lasso Model
from sklearn.linear_model import Lasso
lasso = Lasso(alpha=.01).fit(X_train_scaled, y_train_scaled)

predictions = lasso.predict(X_test_scaled)

MSE = mean_squared_error(y_test_scaled, predictions)
r2 = lasso.score(X_test_scaled, y_test_scaled)
print(f"MSE: {MSE}, R2: {r2}")

```

```python
# Ridge model

from sklearn.linear_model import Ridge
ridge = Ridge(alpha=.01).fit(X_train_scaled, y_train_scaled)

predictions = ridge.predict(X_test_scaled)

MSE = mean_squared_error(y_test_scaled, predictions)
r2 = ridge.score(X_test_scaled, y_test_scaled)
print(f"MSE: {MSE}, R2: {r2}")
```

```python
# ElasticNet model
from sklearn.linear_model import ElasticNet
elasticnet = ElasticNet(alpha=.01).fit(X_train_scaled, y_train_scaled)

predictions = elasticnet.predict(X_test_scaled)

MSE = mean_squared_error(y_test_scaled, predictions)
r2 = elasticnet.score(X_test_scaled, y_test_scaled)
print(f"MSE: {MSE}, R2: {r2}")
```
